```python
# imports

import urllib.request
import string
import nltk
nltk.download('punkt')
nltk.download('stopwords')
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
```

```python
import pandas as pd
df = pd.read_csv('LabeledTweets(2).csv')
df = df.filter(['Content'],axis=1)
df
```

```python
wordList = []
for i,row in df.iterrows():
    tokens = word_tokenize(row['Content'].lower())
    wordList.extend(tokens)
```

```python
len(wordList)
```

```python
wordMap = {}
```

```python
for word in wordList:
    if word not in wordMap:
        wordMap[word] = 1
    else:
        wordMap[word] += 1
```

```python
orderedMap = sorted(wordMap.items(), key = lambda x: x[1], reverse = True)
```

```python
len(orderedMap)
```

```python
orderedMap
```

```python
wordList = list(wordMap.keys())
```

```python
wordList = [word.lower() for word in wordList if (word not in stopwords.words('english')) and (word not in string.punctuation)]
```
