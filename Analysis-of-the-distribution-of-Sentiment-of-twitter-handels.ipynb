```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statistics

df = pd.read_csv('2LabeledScoredTweets.csv')
df = df.drop(['Unnamed: 0','Date'],axis=1)
Handle = df.groupby('Username')

firstSquawk = Handle.get_group('FirstSquawk')

WSJ = Handle.get_group('WSJ')

Business = Handle.get_group('business')

DeltaOne = Handle.get_group('DeItaone')

finaitialJuice = Handle.get_group('financialjuice')

Reuters = Handle.get_group('Reuters')

yahooFinance = Handle.get_group('YahooFinance')

unusualWhales = Handle.get_group('unusual_whales')

zeroHedge = Handle.get_group('zerohedge')

PriapusIQ = Handle.get_group('PriapusIQ')

#######################

Pos = df.loc[df['Score'] > 0]
Pos = Pos.drop(['Content','Target'],axis=1)

Neg = df.loc[df['Score'] < 0]
Neg = Neg.drop(['Content','Target'],axis=1)

```

```python
#firstSquawk

Scores = list(firstSquawk.Score)
std = statistics.stdev(Scores)
mean = statistics.mean(Scores)
Smin = min(Scores)
Smax = max(Scores)

plt.title('Score Distrabution of @FirstSquawk')
plt.hist(Scores, bins=range(Smin,Smax,1), density=True, alpha=0.7, color='blue',edgecolor='black')
plt.axvline(mean, color='red', linestyle='dashed', linewidth=1, label='Mean')
plt.axvline(mean + std, color='green', linestyle='dashed', linewidth=1, label='1 Std')
plt.axvline(mean - std, color='green', linestyle='dashed', linewidth=1)
plt.xlabel('Tweet "Score"')
plt.ylabel('Perentage of Tweets')
plt.legend()
plt.show()
```
