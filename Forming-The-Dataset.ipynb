```python
import pandas as pd

WSJ = pd.read_csv('WSJ.csv')

Reuters = pd.read_csv('Reuters.csv')

Business = pd.read_csv('business.csv')

deltaOne = pd.read_csv('deItaone.csv')

financialJuice = pd.read_csv('financialjuice.csv')

firstSquawk = pd.read_csv('FirstSquawk.csv')

priapusIQ = pd.read_csv('PriapusIQ.csv')

unusualWhales = pd.read_csv('unusualWhales.csv')

zeroHedge = pd.read_csv('zeroHedge.csv')

yahooFinance = pd.read_csv('YahooFinance.csv')


df = pd.merge(WSJ, Reuters, how = 'outer')
df = pd.merge(df, Business, how = 'outer')
df = pd.merge(df, deltaOne, how = 'outer')
df = pd.merge(df, financialJuice, how = 'outer')
df = pd.merge(df, firstSquawk, how = 'outer')
df = pd.merge(df, priapusIQ, how = 'outer')
df = pd.merge(df, unusualWhales, how = 'outer')
df = pd.merge(df, zeroHedge, how = 'outer')
df = pd.merge(df, yahooFinance, how = 'outer')

df = df.drop(['Unnamed: 0'],axis=1)
df['Date'] = df['Date'].str.slice(0,10)
```
